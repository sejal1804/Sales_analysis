## Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Sales = pd.read_csv('sales_data.csv')
```

```python
#Drop the column Region_and_sales_Rep
```

```python
Sales = Sales.drop('Region_and_Sales_Rep' , axis = 1)
```

```python
Sales
```

```python
Sales.head()
```

```python
# Create column Profit
```

```python
Sales['Profit'] = (Sales['Unit_Price'] - Sales['Unit_Cost'])* Sales['Quantity_Sold']
```

```python
Sales
```

```python
# To check sales of each product.
```

```python
plt.figure(figsize=(6, 4))
ax = sns.barplot(x='Product_Category' , y='Sales_Amount' , data = Sales, 
                  hue = 'Product_Category' , legend = False ,palette = ['red' , 'skyblue' , 'green' , 'orange'],
                  errorbar = None, edgecolor = 'k')

for container in ax.containers:
    ax.bar_label(container)
    
plt.title('Products Sales')
plt.xlabel('No of Products')
plt.ylabel(' Sales')
plt.savefig('my_plot1.png')
plt.show()
```

```python
# To check no of products sold in each region.
```

```python
plt.figure(figsize=(6,4))
ax = sns.barplot(x = 'Region' , y = 'Quantity_Sold', data = Sales  ,color = 'purple' ,  errorbar =  None, edgecolor = 'k')

for container in ax.containers:
    ax.bar_label(container)
    
plt.title('No of products sold in a region')
plt.savefig('my_plot2.png')
plt.show()
```

```python
# To check maximum number of payments methods used.
```

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(x = 'Payment_Method' ,data= Sales , edgecolor = 'k')

for container in ax.containers:
    ax.bar_label(container)
plt.title('Count of Payment Methods')
plt.savefig('my_plot3.png')
plt.show()
```

```python
# To check profit of each product.
```

```python
plt.figure(figsize=(6,4))
ax = sns.barplot(x = 'Product_Category' , y = 'Profit' , data = Sales , color = 'gray' , edgecolor = 'k' , errorbar = None)

for container in ax.containers:
    ax.bar_label(container)

plt.title('Product Profit')
plt.savefig('my_plot7.png')
plt.show()
```

```python
# TO check profit by region.
```

```python
plt.figure(figsize=(6,4))
ax = sns.barplot(x = 'Region' , y = 'Profit' , data = Sales , color = 'pink' , edgecolor = 'k' , errorbar = None)

for container in ax.containers:
    ax.bar_label(container)
plt.title('Profit by Region')
plt.savefig('my_plot8.png')
plt.show()
```

```python
# To check Total Quantity bought per Customer Type.
```

```python
plt.figure(figsize=(6,4))
Total_Quantity_bought_by_customer = Sales.groupby('Customer_Type')['Quantity_Sold'].sum().reset_index()
sns.barplot(x = 'Customer_Type' , y = 'Quantity_Sold' , data = Total_Quantity_bought_by_customer ,  hue = 'Customer_Type' , legend = False ,palette = 'pastel' , edgecolor = 'k' )
plt.title('Total Quantity bought per Customer Type')
plt.savefig('my_plot9.png')
plt.show()
```

```python
# To check most preffered Sales Channel.
```

```python
plt.figure(figsize=(6,4))
Sales_Channel = Sales.groupby('Sales_Channel').size()
plt.pie(Sales_Channel.values , labels = Sales_Channel.index , autopct = '%1.1f%%', colors = ['red' , 'green'] , startangle = 140 , wedgeprops = {'edgecolor' : 'k' } )
plt.title('Sales_Channel')
plt.savefig('my_plot4.png')
plt.show()
```

```python
# To check correlation between unit_cost , unit_price , Quantity_sold.
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(Sales[['Quantity_Sold'  , 'Unit_Price' , 'Unit_Cost' ]].corr() , annot = True)
plt.title('Correlation Heatmap of Sales Data')
plt.savefig('my_plot6.png')
plt.show()
```

```python
# To check the Customer_Type which are high in number.
```

```python
plt.figure(figsize=(6,4))
Customer = Sales.groupby('Customer_Type').size()
plt.pie(Customer.values , labels = Customer.index , autopct = '%1.1f%%'  , startangle = 110  , wedgeprops = {'edgecolor' : 'k'})
plt.title('Customer_Type')
plt.savefig('my_plot5.png')
plt.show()
```
